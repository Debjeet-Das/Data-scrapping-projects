```python
# We are going to scrape data from the below link 
url = "https://en.wikipedia.org/wiki/List_of_terrorist_incidents"

# Here the .get function will open a new browser for the given url
browser.get(url)

# Here the .page_source function will copy the source code from the given url
html=browser.page_source

# importing bs4 and using BEautifulSoup to make the extrcated page source parsable using "html.parser"
import bs4
soup=bs4.BeautifulSoup(html,"html.parser")
```

```python
# extracting the links from the web page

# "div" is the tag name and class_ will represent the class name
links=soup.find_all("div",class_="div-col columns column-width")
    
urls=links[0].find_all("ul")
link=[]
for i in urls[0].find_all("li"):
    # this will collect all the links which are updated after 2014 
    if (int(i.a["href"].split("_")[-1])>2014):
        #concatenating the url in the attribute href with the extension to make it a valid link.
        link.append("https://en.wikipedia.org/" + i.a["href"]) 
        
```

```python
# extracted links and stored it in the list link
link
```

```python
# Defining user defined function which will scrape data from the url and and then will return a dataframe with derived columns as year and month

def get_data(x):
    import os
    import pandas as pd
    from selenium import webdriver

    #path="C:\\Users\\debje\\Desktop\\PYTHON"
    #os.chdir(path)
    browser.get(x)
    html = browser.page_source

    import bs4
    soup=bs4.BeautifulSoup(html,"html.parser")
    
    table = soup.find_all("table")
    rows= table[0].find_all("tr")
    
    date=[]
    Type=[]
    Dead=[]
    Injured=[]
    Location=[]
    Details=[]
    per=[]
    part=[]
    for i in rows[1:]:
        date.append(i.find_all("td")[0].text.strip())
        Type.append(i.find_all("td")[1].text.strip())
        Dead.append(i.find_all("td")[2].text.strip())
        Injured.append(i.find_all("td")[3].text.strip())
        Location.append(i.find_all("td")[4].text.strip())
        Details.append(i.find_all("td")[5].text.strip())
        per.append(i.find_all("td")[6].text.strip())
        try:
            part.append(i.find_all("td")[7].text.strip())
        except:
            part.append("data not available")
    import pandas as pd
    data=pd.DataFrame({"Date":date,"Type":Type,"Dead":Dead,"Injured":Injured,"Location":Location,"Details":Details,"Perpetrator":per,"Part":part})
    data["Year"]=x.split("_")[-1] # adding col to main dataframe
    data["Month"]=x.split("_")[-2] # adding col to main dataframe
    return data
```

```python
# Creating dataframes by going to next urls
# here we are also storing the dataframes in the main dictionary with key as page no. and value as dagtaframe
main={}
from tqdm import tqdm
for i in tqdm(range(0,len(link))):
    url5=link[i]
    main[i]=get_data(url5)
      
 
```

```python
# concatenating the DataFrames and creating one big dataframe big_data 
import pandas as pd
big_data=pd.concat([main[i] for i in main],axis=0)
```

```python
big_data=big_data.reset_index()
```
